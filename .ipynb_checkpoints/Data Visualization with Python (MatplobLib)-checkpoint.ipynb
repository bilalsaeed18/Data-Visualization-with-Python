# import pandas Library 

Pandas is a popular Python library used in data analysis and manipulation. It provides easy-to-use data structures and functions for working with structured data, such as tabular data, time series, and more.

```python
import pandas as pd
```

```python
from matplotlib import pyplot as plt
```

```python
x = [10,20,30,40]
y = [12,33,40,55]
z = [80,45,10, 2]

plt.plot(x,y)
plt.plot(x,z)
plt.title("MatplotLib")
plt.xlabel("x")
plt.ylabel("y and z")
plt.legend(["This is Y","This is Z"])
plt.show
```

DataFrame: A DataFrame is a two-dimensional, tabular data structure in Pandas that is similar to a spreadsheet or a SQL table. It consists of rows and columns, where each column can have a different data type. The DataFrame is a fundamental Pandas object and is widely used for data manipulation and analysis.

iloc: iloc is a method in Pandas that is used for integer-location based indexing in a DataFrame. It allows you to select rows and columns from a DataFrame using integer indices

```python
plt.plot(sample_data.column_a, sample_data.column_b, 'o')
plt.plot(sample_data.column_a, sample_data.column_c)
plt.title("Sample Data")
plt.xlabel("x")
plt.ylabel("y and z")
plt.legend(["This is Y","This is Z"])
plt.show()
```

```python
sample_data_2 = pd.read_csv('dataset/countries.csv')
```

```python
# Compare the growth between US and China
```

```python
us = sample_data_2[sample_data_2.country == 'United States']
```

```python
us
```

```python
China = sample_data_2[sample_data_2.country == 'China']
```

```python
China
```

```python
plt.plot(us.year, us.population / 10**6)
plt.plot(China.year, China.population / 10**6)
plt.title("Growth Comaprison between US and China")
plt.xlabel("Year")
plt.ylabel("Population")
plt.legend(["US","China"])
plt.show()
```

```python
us.population
```

```python
us.population / us.population.iloc[0] * 100
```

```python
plt.plot(us.year, us.population / us.population.iloc[0] * 100)
plt.plot(China.year, China.population / China.population.iloc[0] * 100)
plt.title("Growth Comaprison between US and China")
plt.xlabel("Year")
plt.ylabel("Population growth ( first year = 100)")
plt.legend(["US","China"])
plt.show()
```
