# Text

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.title('Axes Title')
plt.xlabel('X Label')
plt.ylabel('Y Label')

plt.text(2,3,'Boxed italic text', style='italic', bbox={'facecolor':'red'})
```

```python
plt.text(2,6,r'an equation: $E = mc^2$', fontsize=20)
```

```python
plt.text(4, 0.05,'color text in axes',verticalalignment ='bottom',color='green' , fontsize=20)
```

```python
plt.plot([2],[1],'o')
plt.show()
```

```python
plt.title('Axes Title')
plt.xlabel('X Label')
plt.ylabel('Y Label')

plt.text(3,8,'Boxed italic text', style='italic', bbox={'facecolor':'red'})
plt.text(2,6,r'an equation: $E = mc^2$', fontsize=20)
plt.text(4, 0.05,'color text in axes',verticalalignment ='bottom',color='green' , fontsize=20)
plt.plot([2],[1],'o')
plt.annotate('annptate', xy=(2,1), xytext=(3,4), arrowprops=dict(facecolor='black', shrink=0.05))

plt.axis([0,10,0,10])
```
