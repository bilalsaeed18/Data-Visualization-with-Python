# Subplot

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.array([0,1,2,4])
y = np.array([3,5,8,10])

plt.subplot(2,3,1)
plt.plot(x,y)



x = np.array([0,1,2,4])
y = np.array([10,15,20,25])

plt.subplot(2,3,2)
plt.plot(x,y)




x = np.array([0,1,2,4])
y = np.array([10,15,20,25])

plt.subplot(2,3,3)
plt.plot(x,y)



x = np.array([0,1,2,4])
y = np.array([10,15,20,25])

plt.subplot(2,3,4)
plt.plot(x,y)



x = np.array([0,1,2,4])
y = np.array([10,15,20,25])

plt.subplot(2,3,5)
plt.plot(x,y)


x = np.array([0,1,2,4])
y = np.array([10,15,20,45])

plt.subplot(2,3,6)
plt.plot(x,y)


plt.show()
```

```python
x = np.array([0,1,2,4])
y = np.array([3,8,5,10])

plt.subplot(2,3,5)
plt.plot(x,y)
plt.title('SALES')


x = np.array([0,1,2,4])
y = np.array([10,15,20,45])

plt.subplot(2,3,6)
plt.plot(x,y)
plt.title('INCOME')


plt.suptitle('My work')
plt.show()
```
