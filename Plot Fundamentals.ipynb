# Line

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
y = np.array([1,8,5,10])


# linestyles are symbol to show how the lines will be displaying
plt.plot(y, linestyle='dotted')
plt.show()
```

```python
plt.plot(y, linestyle='-.', color='g')
plt.show()
```

```python
# we can use hash code for the color
# other parameters can be used such as linewidth

plt.plot(y, linestyle='-.', color='#E12D07', linewidth='5')
plt.show()
```

```python
x = np.array([6,18,15,14])
y = np.array([10,8,15,20])


plt.plot(x)
plt.plot(y)
plt.show()
```

# Markers

```python
y = np.array([2,4,1,7])

plt.plot(y, marker='o')
plt.show()
```

there are lots of symbols we can use to display marker in plot

see documentation here: https://matplotlib.org/3.3.3/api/markers_api.html

```python
y = np.array([2,4,1,7])

# ms = marker size, mec = marker edges color, mfc = marker face color

plt.plot(y, marker='*', ms=10, mec='#E12D07', mfc='green')
plt.show()
```

#### There is a format for plots called Marker|Line|color

```python
y = np.array([2,4,1,7])

plt.plot(y, 'o:g' )
plt.show()
```

```python
n = 256
x = np.linspace(-np.pi, np.pi, n, endpoint=True)
y = np.sin(2*x)

plt.plot(x,y, color='blue', alpha=1.00)
plt.fill_between(x,y,1,color='blue', alpha=.1)

plt.show()
```

# Legend Label Title

```python
a = np.array([0,1,2,3])
b = np.array([3,8,1,9])
x = np.array([0,1,2,3])
y = np.array([6,2,7,11])

plt.plot(a,b,x,y)
plt.show()
```

```python
a = np.array([0,1,2,3])
b = np.array([3,8,1,9])
x = np.array([0,1,2,3])
y = np.array([6,2,7,11])

plt.plot(a,b,x,y)
plt.xlabel('Profit')
plt.ylabel('Month')
plt.title('Monthly Profit')
plt.legend(labels=('May','April'), loc='lower right')
plt.show()
```

```python
a = np.array([0,1,2,3])
b = np.array([3,8,1,9])
x = np.array([0,1,2,3])
y = np.array([6,2,7,11])

plt.plot(a,b,x,y, marker='o')
plt.xlabel('Profit')
plt.ylabel('Month')
plt.title('Monthly Profit')
plt.legend(labels=('May','April'), loc='lower right')
plt.show()
```

# Grid

```python
y = [1, 4, 9, 16, 25, 36]
z = [1, 16, 42, 55, 70, 90]
x = [1, 6, 12, 18, 24, 48]

plt.plot(z,y,'ys-')
plt.plot(x,y, 'go--')
plt.legend(labels=('TV', 'Telephone'))
plt.grid(True)
plt.title('Ad effects on Sales')
plt.xlabel('Medium')
plt.ylabel('Sales')
plt.show()
```

```python
y = [1, 4, 9, 16, 25, 36]
z = [1, 16, 42, 55, 70, 90]
x = [1, 6, 12, 18, 24, 48]

plt.plot(z,y,'ys-')
plt.plot(x,y, 'go--')
plt.legend(labels=('TV', 'Telephone'))
plt.grid(color='blue', ls='-', lw='0.25')
plt.title('Ad effects on Sales')
plt.xlabel('Medium')
plt.ylabel('Sales')
plt.show()
```
