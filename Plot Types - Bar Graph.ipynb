# Bar Graph

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.array(['A','B','X','Y'])
y = np.array([3,9,2,11])

plt.bar(x,y)
plt.show()
```

```python
x = np.array(['Apple','Banana'])
y = np.array([270,150])

plt.bar(x,y)
plt.show()
```

```python
x = np.array(['A','B','X','Y'])
y = np.array([3,9,2,11])

plt.barh(x,y)
plt.show()
```

```python
x = np.array(['A','B','X','Y'])
y = np.array([3,9,2,11])

plt.bar(x,y, color='#000000')
plt.show()
```

```python
plt.bar(x,y, color='red')
plt.show()
```

```python
x = np.array(['A','B','X','Y'])
y = np.array([3,9,2,11])

plt.bar(x,y, color='#4CAF50', width=0.5)
plt.show()
```

```python
x = np.array(['A','B','X','Y'])
y = np.array([3,9,2,11])

plt.barh(x,y, color='#4CAF50', height=0.5)
plt.show()
```

```python
Men = [20,35,30,32,27]
Women = [19,32,24,21,25]
group = np.arange(5)

plt.bar(group, Men, width=.35, color='g')
plt.bar(group, Women, width=.35, bottom=Women, color='b')


plt.xlabel('Scores')
plt.ylabel('Groups')
plt.title('Scores by Group and Gender')
plt.xticks(group,('G1','G2','G3','G4','G5'))
plt.yticks(np.arange(0, 81, 10))
plt.legend(labels=['Men','Women'])

plt.show()
```

```python
lans = np.array(['Python','C++','Php','Java'])
students = np.array([23,17,40,28])

plt.bar(lans, students)
plt.show()
```
