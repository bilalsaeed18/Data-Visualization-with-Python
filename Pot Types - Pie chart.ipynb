# Pie Chart

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
y = np.array([20,35,30,15])
lbs =['Apple','Banana','Cherry','Fruit']

plt.pie(y, labels=lbs)
plt.show()
```

```python
y = np.array([20,35,30,15])
lbs =['Apple','Banana','Cherry','Fruit']
exp = [0.2,0,0,0]

plt.pie(y, labels=lbs, startangle=180, explode=exp)
plt.show()
```
