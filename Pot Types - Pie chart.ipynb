# Pie Chart

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
y = np.array([20,35,30,15])
lbs =['Apple','Banana','Cherry','Fruit']

plt.pie(y, labels=lbs)
plt.show()
```

```python
y = np.array([20,35,30,15])
lbs =['Apple','Banana','Cherry','Fruit']
exp = [0.2,0,0,0]

plt.pie(y, labels=lbs, startangle=180, explode=exp, shadow=True)
plt.show()
```

```python
y = np.array([20,35,30,15])
lbs =['Apple','Banana','Cherry','Fruit']
exp = [0.2,0,0,0]

plt.pie(y, labels=lbs)
plt.legend(title='Four Fruits', loc='upper left')
plt.show()
```

```python
y = np.array([20,35,30,15])
lbs =['Apple','Banana','Cherry','Fruit']

plt.pie(y)
plt.legend(labels=lbs, title='Four Fruits')
plt.show()
```

```python
lans = np.array(['Python','C++','Php','Java'])
students = np.array([23,17,40,28])

plt.pie(students, labels=lans, autopct='%1.2f%%')
plt.show()
```
